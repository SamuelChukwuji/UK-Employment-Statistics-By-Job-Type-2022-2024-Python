```python
#Reducing the dataset to periods of 2022 and upwards
df = df[df['Period'].str.contains('2022|2023|2024')]
df
```

```python
#converting the object columns to numeric and removing the commas
df['Total in Employment'] = df['Total in Employment'].str.replace(',', '').astype(float)
df['Employees'] = df['Employees'].str.replace(',', '').astype(float)
df['Self employed'] = df['Self employed'].str.replace(',', '').astype(float)
df['Total people working full-time'] = df['Total people working full-time'].str.replace(',', '').astype(float)
df['Total people working part-time'] = df['Total people working part-time'].str.replace(',', '').astype(float)
df['Employees working full-time'] = df['Employees working full-time'].str.replace(',', '').astype(float)
df['Employees working part-time'] = df['Employees working part-time'].str.replace(',', '').astype(float)
df['Self-employed people working full-time'] = df['Self-employed people working full-time'].str.replace(',', '').astype(float)
df['Self-employed people working part-time'] = df['Self-employed people working part-time'].str.replace(',', '').astype(float)
df['Total workers with second jobs'] = df['Total workers with second jobs'].str.replace(',', '').astype(float)
df['Total in Temporal Employment'] = df['Total in Temporal Employment'].str.replace(',', '').astype(float)
df['Total-Part-time'] = df['Total- Part-time'].str.replace(',', '').astype(float)
df['Part-time-Could not find full-time job'] = df['Part-time-Could not find full-time job'].str.replace(',', '').astype(float)
df['Part-time-Did not want full-time job'] = df['Part-time-Did not want full-time job'].str.replace(',', '').astype(float)
df['Part-time- Student or at school'] = df['Part-time- Student or at school'].str.replace(',', '').astype(float)
```

```python
def custom_date_parser(date_str):
    if 'Nov-Jan 2022' in date_str:
        return '2022-01-01'  
    elif 'Dec-Feb 2022' in date_str:
        return '2022-02-01'  
    elif 'Jan-Mar 2022' in date_str:
        return '2022-03-01'
    elif 'Feb-Apr 2022' in date_str:
        return '2022-04-01'
    elif 'Mar-May 2022' in date_str:
        return '2022-05-01'
    elif 'Apr-Jun 2022' in date_str:
        return '2022-06-01'
    elif 'May-Jul 2022' in date_str:
        return '2022-07-01'
    elif 'Jun-Aug 2022' in date_str:
        return '2022-08-01'
    elif 'Jul-Sep 20225' in date_str:
        return '2022-09-01'
    elif 'Aug-Oct 2022' in date_str:
        return '2022-10-01'
    elif 'Sep-Nov 2022' in date_str:
        return '2022-11-01'  
    elif 'Oct-Dec 2022' in date_str:
        return '2022-12-01'
    elif 'Nov-Jan 2023' in date_str:
        return '2023-01-01'
    elif 'Dec-Feb 2023' in date_str:
        return '2023-02-01'
    elif 'Jan-Mar 2023' in date_str:
        return '2023-03-01'
    elif 'Feb-Apr 2023' in date_str:
        return '2023-04-01'
    elif 'Mar-May 2023' in date_str:
        return '2023-05-01'
    elif 'Apr-Jun 2023' in date_str:
        return '2023-06-01'
    elif 'May-Jul 2023' in date_str:
        return '2023-07-01'
    elif 'Jun-Aug 2023' in date_str:
        return '2023-08-01'
    elif 'Jul-Sep 2023' in date_str:
        return '2023-09-01'
    elif 'Aug-Oct 2023' in date_str:
        return '2023-10-01'
    elif 'Sep-Nov 2023' in date_str:
        return '2023-11-01'  
    elif 'Oct-Dec 2023' in date_str:
        return '2023-12-01'
    elif 'Nov-Jan 2024' in date_str:
        return '2024-01-01'
    elif 'Dec-Feb 2024' in date_str:
        return '2024-02-01'
    elif 'Jan-Mar 2024' in date_str:
        return '2024-03-01'
    elif 'Feb-Apr 2024' in date_str:
        return '2024-04-01'
    elif 'Mar-May 2024' in date_str:
        return '2024-05-01'
    else:
        return date_str 
```

```python
df['Period'] = df['Period'].apply(custom_date_parser)
```

```python
#converting the period column to datetime
df['Period'] = pd.to_datetime(df['Period'], errors='coerce')
```
